Problem - 1:
Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.
Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,
median, maximum value, and minimum value of the ’Fare’ column?

```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df.head()
```

```python
filtered_df = df[df['Pclass'] == 1]

selected_column = ['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']
new_df = filtered_df[selected_column]
```

```python
new_df.head()
```

```python
 new_df.shape
```

```python
print(f"Min: {new_df['Fare'].min()}")
print(f"Max: {new_df['Fare'].max()}")
print(f"Mean: {new_df['Fare'].mean()}")
print(f"Median: {new_df['Fare'].median()}")
```

```python
df['Embarked'] = df['Embarked'].replace({
    'C': 'Cherbourg',
    'Q': 'Queenstown',
    'S': 'Southampton'
})
```

```python
print(df['Embarked'])
```

```python
df.head()
```

```python
encoded_columns = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, encoded_columns], axis=1)
df = df.drop(columns=['Embarked'])
```

```python
df.head()
```

```python
mean_survival_rates = df.groupby('Sex')['Survived'].mean()
print("Mean Survival Rates by Gender:")
print(mean_survival_rates)
```

```python

# Bar plot for mean survival rates
plt.subplot(1, 2, 1)
sns.barplot(x=mean_survival_rates.index, y=mean_survival_rates.values)
plt.title('Mean Survival Rates by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Gender')

# Box plot for survival distribution by gender
plt.subplot(1, 2, 2)
sns.boxplot(x='Sex', y='Survived', data=df)
plt.title('Survival Distribution by Gender')
plt.ylabel('Survival')
plt.xlabel('Gender')

plt.tight_layout()
plt.show()
```

```python
null_df = df.isnull().sum()
print(null_df)
```

```python
g = sns.catplot(
    data=df, 
    x='Sex', 
    y='Survived', 
    hue='Sex', 
    kind='bar', 
    palette='pastel',
    height=5, 
    aspect=1
)

# Customize the titles and axes
g.set_titles('{col_name}')
g.set_axis_labels('Gender', 'Mean Survival Rate')
g.set_xticklabels(['Female', 'Male'])
g.fig.suptitle('Survival Rates by Gender and Port of Embarkation', y=1.02)

# Show plot
plt.show()
```
